```python
def get_CES_2goods_equation(
    out: str, fac1: str, fac2: str, p1: str, p2: str, p_out: str, A: str, alpha: str, epsilon: str
):

    eq_production = Template(
        f"$out = $A * ($alpha * $fac1 ** (($epsilon - 1) / $epsilon) +"
        + f"(1 - $alpha) * $fac2 ** (($epsilon - 1) / $epsilon)) ** ($epsilon / ($epsilon - 1))"
    ).safe_substitute(locals())

    eq_fac1_demand = Template(
        f"$fac1 = $out / $A * ($alpha * $p_out * $A / $p1) ** $epsilon"
    ).safe_substitute(locals())

    eq_fac2_demand = Template(
        f"$fac2 = $out / $A * ((1 - $alpha) * $A * $p_out / $p2) ** $epsilon"
    ).safe_substitute(locals())

    return eq_production, eq_fac1_demand, eq_fac2_demand
```

```python
def get_CES_2goods_params(fac1, fac2, p1, p2, output, epsilon):
    _rho = (epsilon - 1) / epsilon
    alpha = p1 * fac1 ** (1 / epsilon) / ((p1 * fac1 ** (1 / epsilon) + p2 * fac2 ** (1 / epsilon)))
    A = output * (alpha * fac1**_rho + (1 - alpha) * fac2**_rho) ** (-1 / _rho)
    return A, alpha


def get_CES_Ngoods_params(facs, ps, output, epsilon):
    _rho = (epsilon - 1) / epsilon
    alpha = ps * facs ** (1 / epsilon) / (ps * facs ** (1 / epsilon)).sum()
    A = output * (alpha * facs**_rho).sum() ** (-1 / _rho)
    return A, alpha


def get_DXS_params(P: np.array, X: np.array, epsilon: float):
    assert P.shape == X.shape, "Price and good shapes do not match"

    _rho = (epsilon - 1) / epsilon
    alpha = P * X ** (1 / epsilon) / (P * X ** (1 / epsilon)).sum()
    A = X.sum() * (alpha * X**_rho).sum() ** (-1 / _rho)
    return A, alpha


def get_CD_params(output: float, output_price: float, factors: np.array, factor_prices: list):
    alpha = factor_prices * factors / (output_price * output)
    A = output / (factors**alpha).prod()
    return A, alpha
```

```python
# inverse parametrisation functions
def get_CES_params_inv(A, alpha, epsilon, p1, p2, output):
    _rho = (epsilon - 1) / epsilon
    _delta = (((1 - alpha) / alpha) * (p1 / p2)) ** epsilon

    f1 = output / A * (1 / (alpha + (1 - alpha) * _delta**_rho)) ** (1 / _rho)
    f2 = _delta * f1

    return f1, f2


def get_Leontief_share(psi, output: float):
    shares = psi * output
    #     total = shares.sum()
    return shares


def get_CD_params_inv(
    A: float, alpha: np.array, factor_prices: np.array, output: float, output_price: float
):
    assert alpha.shape == factor_prices.shape, "Inhomogeneous dimensions"

    # if not output is provided and only scale (e.g. initial production function)
    #     output = (A * (alpha**alpha).prod())**(1 / (1 - alpha.sum())) * scale

    factors = (alpha * output * output_price) / factor_prices
    return factors
```

```python
def calibrate_model_inv(
    scale,
    tau_income,
    tau_sales,
    tau_capital,
    tau_wage,
    psi_VA,
    psi_VC,
    psi_X,
    A,
    alpha,
    alpha_C,
    alpha_I,
    alpha_G,
    epsilon,
    mpc,
    I_G,
    L_s,
    K_s,
    P_bar,
):

    Y = scale
    # Normalized prices
    w = 1.0
    resid = 0.0
    r = 1.0

    P_VA = np.ones(n_sectors)
    P_VC = np.ones(n_sectors)
    P_Y = np.ones(n_sectors)

    P_VC_E = np.ones(n_eprod)
    P_Y_E = np.ones(n_eprod)

    P_KE_S = np.ones(n_sectors)
    P_KL_E = np.ones(n_eprod)

    P_W = np.ones(n_eprod)
    P_R = np.ones(n_eprod)

    P = P_Y * (1 + tau_sales)
    r_star = r * (1 + tau_capital)
    w_star = w * (1 + tau_wage)

    # Numeraire
    P_bar = P[0]

    # value chains
    VA = psi_VA * Y
    VC = psi_VC * Y
    VC_E = psi_VC_E * Y_E
    X = psi_X * VC[None]
    X_E = psi_X_E * VC_E[None]

    # Energy mix aggregator
    A_Es, alpha_Es = get_CES_Ngoods_params(Y_E, P_EE, E_s, epsilon_Es)

    # Firm calibration
    KE_S_d, L_S_d = get_CES_params_inv(A_VA, alpha_VA, epsilon_VA, P_KE_S, w_star, VA)
    K_S_d, E_d = get_CES_params_inv(A_KE_S, alpha_KE_S, epsilon_KE_S, r_star, P_E, KE_S_d)

    K_E_d, L_E_d = get_CES_params_inv(A_KL_E, alpha_KL_E, epsilon_KL_E, r_star, w_star, KL_E_d)
    KL_E_d, R_d = get_CES_params_inv(A_W, alpha_W, epsilon_W, P_KL_E, P_R, W_s)

    # Factor supplies
    L_s = L_S_d.sum() + L_E_d.sum()
    K_s = K_S_d.sum() + K_E_d.sum()
    E_s = E_d.sum()
    R_s = R_d

    U = np.prod(C**alpha_C)

    # income
    income = r * K_s + w * L_s + (P_R * R_d).sum()
    income_tax_payment = tau_income * income
    net_income = (1 - tau_income) * income

    # consumption
    S = (1 - mpc) * net_income
    consumption_spend = mpc * net_income
    C = alpha_C * consumption_spend / P
    U = (C**alpha_C).prod()

    Gamma = tau_income * income + (
        tau_sales * P_Y * C + tau_capital * r * K_d + tau_wage * w * L_d
    ).sum(axis=0)
    G = Gamma - S_G
    C_G = G * alpha_G

    I_s = S + S_G
    I_d = alpha_I * I_s

    ## figure out tax on K_E_d, K_S_d
    # Tax Rates
    sales_tax_payment = tau_sales * C
    capital_tax_payment = tau_capital * K_d
    wage_tax_payment = tau_wage * L_d

    d = {}
    for obj in mod.variables + mod.parameters:
        d[obj.name] = deepcopy(locals()[obj.name])

    return d
```

```python
def calibrate_model(
    sales_tax_payment,
    income_tax_payment,
    capital_tax_payment,
    wage_tax_payment,
    S,
    S_G,
    L_S_d,
    K_S_d,
    L_E_d,
    K_E_d,
    E_d,
    R_d,
    C,
    C_G,
    Y,
    Y_E,
    X,
    X_E,
    I_d,
    epsilon_VA,
    epsilon_KE_S,
    epsilon_KL_E,
    epsilon_W,
    epsilon_Es,
    variables,
    parameters,
):

    # Normalize prices to 1
    w = 1.0
    r = 1.0
    P_E = 1.0

    P_VA = np.ones(n_sectors)
    P_VC = np.ones(n_sectors)
    P_Y = np.ones(n_sectors)

    P_VC_E = np.ones(n_eprod)
    P_Y_E = np.ones(n_eprod)

    P_KE_S = np.ones(n_sectors)
    P_KL_E = np.ones(n_eprod)

    P_W = np.ones(n_eprod)
    P_R = np.ones(n_eprod)

    # taxes
    tau_sales = sales_tax_payment / C
    tau_capital = capital_tax_payment / K_d
    tau_wage = wage_tax_payment / L_d

    P = P_Y * (1 + tau_sales)
    r_star = r * (1 + tau_capital)
    w_star = w * (1 + tau_wage)

    # Zero the residual
    resid = 0.0

    # Factor supplies
    L_s = L_S_d.sum() + L_E_d.sum()
    K_s = K_S_d.sum() + K_E_d.sum()
    E_s = E_d.sum()
    R_s = R_d

    # Numeraire
    P_num = P[0]

    X /= P[:, None]
    X_E /= P[:, None]
    C /= P
    C_G /= P
    I_d /= P
    Y /= P
    Y_E /= P_EE

    # Firm calibration
    KE_S_d = (K_S_d * r + E_d * P_E) / P_KE_S
    KL_E_d = (K_E_d * r + L_E_d * w) / P_KL_E
    W_s = (KL_E_d * P_KL_E + R_d * P_R) / P_W
    VA = (P_KE_S * KE_S_d + w * L_S_d) / P_VA

    VC = (P[:, None] * X).sum(axis=0) / P_VC
    VC_E = (P[:, None] * X_E).sum(axis=0) / P_VC_E

    # Final good sector Value Add
    A_VA, alpha_VA = get_CES_2goods_params(KE_S_d, L_S_d, P_KE_S, w, VA, epsilon_VA)
    A_KE_S, alpha_KE_S = get_CES_2goods_params(K_S_d, E_d, r, P_E, KE_S_d, epsilon_KE_S)

    # Energy sector production function
    A_KL_E, alpha_KL_E = get_CES_2goods_params(K_E_d, L_E_d, r, w, KL_E_d, epsilon_KL_E)
    A_W, alpha_W = get_CES_2goods_params(KL_E_d, R_d, P_KL_E, P_R, W_s, epsilon_W)

    # Energy mix aggregator
    A_Es, alpha_Es = get_CES_Ngoods_params(Y_E, P_EE, E_s, epsilon_Es)

    psi_VA = VA / Y
    psi_VC = VC / Y
    psi_VC_E = VC_E / Y_E
    psi_X = X / VC[None]
    psi_X_E = X_E / VC_E[None]
    psi_W = W_s / Y_E

    income = w * L_s + r * K_s + (P_R * R_d).sum()
    tau_income = income_tax_payment / income
    net_income = (1 - tau_income) * income

    mpc = 1 - S / net_income
    consumption_spend = mpc * net_income

    alpha_C = P * C / consumption_spend
    U = np.prod(C**alpha_C)

    # Exogenous government spending level
    S_G_bar = S_G

    G = C_G.sum()
    alpha_G = C_G / G

    I_s = I_d.sum()
    alpha_k = I_d / I_s

    d = {}
    for obj in variables + parameters:
        if obj.name != "_":
            d[obj.name] = locals()[obj.name]

    return d
```

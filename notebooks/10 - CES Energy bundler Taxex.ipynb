```python
# inverse parametrisation functions
def get_CES_params_inv(A, alpha, epsilon, p1, p2, p_out, output):
    _rho = (epsilon - 1) / epsilon

    f1 = output / A * (alpha * p_out * A / p1) ** epsilon
    f2 = output / A * ((1 - alpha) * p_out * A / p2) ** epsilon

    return f1, f2
```

```python
def calibrate_model_inv(
    Y,
    alpha_VA,
    alpha_KE_S,
    alpha_KL_E,
    alpha_W,
    alpha_k,
    alpha_Es,
    A_VA,
    A_W,
    A_KE_S,
    A_KL_E,
    A_Es,
    psi_VA,
    psi_VC,
    psi_X,
    psi_VC_E,
    psi_X_E,
    psi_W,
    epsilon_VA,
    epsilon_W,
    epsilon_KE_S,
    epsilon_KL_E,
    epsilon_Es,
    tau_income,
    tau_sales,
    tau_wage,
    tau_capital,
    alpha_C,
    mpc,
    alpha_G,
    L_s,
    R_s,
    K_s,
    P_num,
    S_G_bar,
):

    assert Y.shape[0] == len(mod.coords["i"]), "Provide output by sector"

    # Normalized prices
    w = 1.0
    resid = 0.0
    r = 1.0
    P_E = 1.0

    P_VA = np.ones(n_sectors)
    P_VC = np.ones(n_sectors)
    P_Y = np.ones(n_sectors)

    P_VC_E = np.ones(n_eprod)
    P_Y_E = np.ones(n_eprod)

    P_KE_S = np.ones(n_sectors)
    P_KL_E = np.ones(n_eprod)

    P_W = np.ones(n_eprod)
    P_R = np.ones(n_eprod)

    P = P_Y * (1 + tau_sales)
    r_star = r * (1 + tau_capital)
    w_star = w * (1 + tau_wage)

    # Numeraire
    P_num = P[0]

    # value chains
    VA = psi_VA * Y
    VC = psi_VC * Y
    X = psi_X * VC[None]

    X = psi_X * (psi_VC * Y)
    VA

    # Firm calibration
    KE_S_d, L_S_d = get_CES_params_inv(
        A=A_VA, alpha=alpha_VA, epsilon=epsilon_VA, p1=P_KE_S, p2=w_star, p_out=P_VA, output=VA
    )
    K_S_d, E_d = get_CES_params_inv(
        A=A_KE_S,
        alpha=alpha_KE_S,
        epsilon=epsilon_KE_S,
        p1=r_star,
        p2=P_E,
        p_out=P_KE_S,
        output=KE_S_d,
    )

    # energy
    E_s = E_d.sum()
    Y_E = E_s / A_Es * (alpha_Es * P_E * A_Es / P_Y_E) ** epsilon_Es

    VC_E = psi_VC_E * Y_E
    X_E = psi_X_E * VC_E[None]
    W_s = psi_W * Y_E

    KL_E_d, R_d = get_CES_params_inv(
        A=A_W, alpha=alpha_W, epsilon=epsilon_W, p1=P_KL_E, p2=P_R, p_out=P_W, output=W_s
    )
    K_E_d, L_E_d = get_CES_params_inv(
        A=A_KL_E, alpha=alpha_KL_E, epsilon=epsilon_KL_E, p1=r, p2=w, p_out=P_KL_E, output=KL_E_d
    )

    # Factor supplies
    L_s = L_S_d.sum() + L_E_d.sum()
    K_s = K_S_d.sum() + K_E_d.sum()
    E_s = E_d.sum()
    R_s = R_d

    # income
    income = r * K_s + w * L_s + (P_R * R_d).sum()
    income_tax_payment = tau_income * income
    net_income = (1 - tau_income) * income

    # consumption
    S = (1 - mpc) * net_income
    consumption_spend = mpc * net_income

    C = alpha_C * consumption_spend / P
    U = (C**alpha_C).prod()

    # government income
    Gamma = tau_income * income + (
        tau_sales * P_Y * C + tau_capital * r * K_S_d + tau_wage * w * L_S_d
    ).sum(axis=0)

    S_G = S_G_bar

    G = Gamma - S_G
    C_G = alpha_G * G

    I_s = S + S_G
    I_d = alpha_k * I_s

    # Tax Rates
    sales_tax_payment = tau_sales * C
    capital_tax_payment = tau_capital * K_s
    wage_tax_payment = tau_wage * L_s

    print(C + C_G + I_d + X_E.sum(axis=1) + X.sum(axis=1), Y)

    d = {}
    for obj in mod.variables + mod.parameters:
        d[obj.name] = deepcopy(locals()[obj.name])

    return d
```

```python
def calibrate_model(
    sales_tax_payment,
    income_tax_payment,
    capital_tax_payment,
    wage_tax_payment,
    S,
    S_G,
    L_S_d,
    K_S_d,
    L_E_d,
    K_E_d,
    E_d,
    R_d,
    C,
    C_G,
    Y,
    Y_E,
    X,
    X_E,
    I_d,
    epsilon_VA,
    epsilon_KE_S,
    epsilon_KL_E,
    epsilon_W,
    epsilon_Es,
    variables,
    parameters,
):

    # Normalize prices to 1
    w = 1.0
    r = 1.0
    P_E = 1.0

    P_VA = np.ones(n_sectors)
    P_VC = np.ones(n_sectors)
    P_Y = np.ones(n_sectors)

    P_VC_E = np.ones(n_eprod)
    P_Y_E = np.ones(n_eprod)

    P_KE_S = np.ones(n_sectors)
    P_KL_E = np.ones(n_eprod)

    P_W = np.ones(n_eprod)
    P_R = np.ones(n_eprod)

    # taxes
    tau_sales = sales_tax_payment / C
    tau_capital = capital_tax_payment / K_d
    tau_wage = wage_tax_payment / L_d

    P = P_Y * (1 + tau_sales)
    r_star = r * (1 + tau_capital)
    w_star = w * (1 + tau_wage)

    # Zero the residual
    resid = 0.0

    # Factor supplies
    L_s = L_S_d.sum() + L_E_d.sum()
    K_s = K_S_d.sum() + K_E_d.sum()
    E_s = E_d.sum()
    R_s = R_d

    # Numeraire
    P_num = P[0]

    X /= P[:, None]
    X_E /= P[:, None]
    C /= P
    C_G /= P
    I_d /= P
    Y /= P
    Y_E /= P_Y_E

    # Firm calibration
    KE_S_d = (K_S_d * r + E_d * P_E) / P_KE_S
    KL_E_d = (K_E_d * r + L_E_d * w) / P_KL_E
    W_s = (KL_E_d * P_KL_E + R_d * P_R) / P_W
    VA = (P_KE_S * KE_S_d + w * L_S_d) / P_VA

    VC = (P[:, None] * X).sum(axis=0) / P_VC
    VC_E = (P[:, None] * X_E).sum(axis=0) / P_VC_E

    # Final good sector Value Add
    A_VA, alpha_VA = get_CES_2goods_params(KE_S_d, L_S_d, P_KE_S, w, VA, epsilon_VA)
    A_KE_S, alpha_KE_S = get_CES_2goods_params(K_S_d, E_d, r, P_E, KE_S_d, epsilon_KE_S)

    # Energy sector production function
    A_KL_E, alpha_KL_E = get_CES_2goods_params(K_E_d, L_E_d, r, w, KL_E_d, epsilon_KL_E)
    A_W, alpha_W = get_CES_2goods_params(KL_E_d, R_d, P_KL_E, P_R, W_s, epsilon_W)

    # Energy mix aggregator
    A_Es, alpha_Es = get_CES_Ngoods_params(Y_E, P_Y_EE, E_s, epsilon_Es)

    psi_VA = VA / Y
    psi_VC = VC / Y
    psi_VC_E = VC_E / Y_E
    psi_X = X / VC[None]
    psi_X_E = X_E / VC_E[None]
    psi_W = W_s / Y_E

    income = w * L_s + r * K_s + (P_R * R_d).sum()
    tau_income = income_tax_payment / income
    net_income = (1 - tau_income) * income

    mpc = 1 - S / net_income
    consumption_spend = mpc * net_income

    alpha_C = P * C / consumption_spend
    U = np.prod(C**alpha_C)

    # Exogenous government spending level
    S_G_bar = S_G

    G = C_G.sum()
    alpha_G = C_G / G

    I_s = I_d.sum()
    alpha_k = I_d / I_s

    d = {}
    for obj in variables + parameters:
        if obj.name != "_":
            d[obj.name] = locals()[obj.name]

    return d
```
